```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Imported CSV to Jupyter notebook
#Removed index column from CSV
#Displayed head to see if it was pulled correctly
graduate_data_df = pd.read_csv('Data/Graduate_Data.csv',index_col=0)
graduate_data_df.head()
```

```python
#Check to see if all data is categorized correctly
graduate_data_df.dtypes
```

```python
#Find all unique values of majors to see if there are any null entries
unique_majors = graduate_data_df['Education.Major'].unique()
print(f'There are {len(unique_majors)} unique majors in this data set')
unique_majors
```

```python
#Dictionary to hold unique data frames
filtered_dfs = {}

# Filter and store the DataFrame for each major
for major in unique_majors:
    filtered_dfs[major] = graduate_data_df[graduate_data_df['Education.Major'] == major]
```

```python
#Graphs for each major based off of salary mean
for major, data in filtered_dfs.items():
    plt.figure(figsize=(5, 5))
    plt.plot(data['Year'].astype(str), data['Salaries.Mean'], marker='o')
    plt.title(f'Mean Salaries Over Years for {major}')
    plt.xlabel('Year')
    plt.ylabel('Mean Salary')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```
